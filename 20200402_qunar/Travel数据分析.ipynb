### 5.去除错误数据

```python
data=data[~data['地点'].isin(['攻略'])]
data=data[~data['天数'].isin(['99+'])]
data
```

### 6.转换数据类型

```python
data['天数']=data['天数'].astype(int)
data['人均费用']=data['人均费用'].astype(int)
```

### 7.筛选数据

```python
data=data.reset_index(drop=True)
data
```

### 8.处理月份

```python
def Month(e):
    
    m=str(e).split('/')[2]
    
    if m=='01':
        return '一月'
    if m=='02':
        return '二月'
    if m=='03':
        return '三月'
    if m=='04':
        return '四月'
    if m=='05':
        return '五月'
    if m=='06':
        return '六月'
    if m=='07':
        return '七月'
    if m=='08':
        return '八月'
    if m=='09':
        return '九月'
    if m=='10':
        return '十月'
    if m=='11':
        return '十一月'
    if m=='12':
        return '十二月'
    
```

```python
data['旅行月份']=data['出发时间'].apply(Month)
data['出发时间']=pd.to_datetime(data['出发时间'])
data
```

### 9.处理浏览量

```python
import re

def Look(e):
    
    if '万' in e:
        num=re.findall('(.*?)万',e)
        return float(num[0])*10000
    else:
        return float(e)
```

```python
data['浏览次数']=data['浏览量'].apply(Look)
data.drop(['浏览量'],axis=1,inplace=True)
data['浏览次数']=data['浏览次数'].astype(int)
data.head()
```

```python
import jieba
import jieba.analyse
import re
```

```python
punc = '~`!#$%^&*()_+-=|\';":/.,?><~·！@#￥%……&*（）——+-=“：’；、。，？》《{}【】'
def remove_fuhao(e):
    short = re.sub(r"[%s]+" % punc, " ", e)
    return short
def cut_word(text):
    text = jieba.cut_for_search(str(text))
    return ' '.join(text)
```

```python
data2=data1
```

```python
data2['简介'] = data2['短评'].apply(remove_fuhao).apply(cut_word)
data2.head()
```

```python
word = data2['简介'].values.tolist()
fb = open(r'.\travel_text.txt','w',encoding='utf-8')
for i in range(len(word)):
    fb.write(word[i])
with open(r'.\travel_text.txt','r',encoding='utf-8')as f:
    words = f.read()
    f.close
```

```python
jieba.analyse.set_stop_words(r'./cn_stopwords.txt')
new_words = jieba.analyse.textrank(words, topK=30, withWeight=True)
print(new_words)
```

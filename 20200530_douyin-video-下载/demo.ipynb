```python
session =requests.session()
response = session.get(url,headers=headers,params=params)
```

```python
data = response.text
pattern = re.compile('"(https://aweme.snssdk.com/aweme/v1/play/.*?)"')
result = pattern.findall(data)
```

```python
result = [i.split("&ratio")[0] for i in result]
result2 = [i.replace("/play/", "/playwm/") for i in result]
```

```python
if not os.path.exists("no_water_mark"):
    os.mkdir("no_water_mark")
```

```python
count = 0
for i in set(result):
    count += 1
    video = session.get(i,headers=headers)
    with open(f"./no_water_mark/{count}.mp4","wb") as f:
        f.write(video.content)
```
